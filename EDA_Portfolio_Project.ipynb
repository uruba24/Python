# **Step 6: Outlier Detection**

```python
# Detect outliers for 'Income' using the IQR method
for column in ['Age', 'Income', 'Miles','Education', 'Usage', 'Fitness']:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify and print outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    print(f"Outliers in {column}:")
    print(outliers)

    # Plot a box plot for visual inspection
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=data[column], color='darkblue')
    plt.title(f'Box Plot for {column}')
    plt.show()
```

# **Step 7: Conditional Probabilities**

**Step 1: Calculate Percentages for Product Categories**

```python
#Calculate the percentage of customers purchasing each treadmill model.
product_counts = data['Product'].value_counts(normalize=True) * 100
print(product_counts)
```

**Step 2: Create Frequency Tables**

```python

# Total number of customers
total_customers = data.shape[0]

# Initialize dictionary for frequency tables
frequency_tables = {}

# Product - Gender
gender_counts = data['Gender'].value_counts()
frequency_tables['Gender'] = pd.DataFrame({
    'Gender': gender_counts.index,
    'Count': gender_counts.values,
    'Percentage': (gender_counts.values / total_customers) * 100
})

# Frequency of Gender for KP781
kp781_gender_counts = data[data['Product'] == 'KP781']['Gender'].value_counts()

# Frequency table for KP781 Gender
frequency_tables['Product'] = pd.DataFrame({
    'Gender': kp781_gender_counts.index,
    'Count': kp781_gender_counts.values,
    'Percentage': (kp781_gender_counts.values / data.shape[0]) * 100
})

# Frequency of Gender for KP281
kp281_gender_counts = data[data['Product'] == 'KP281']['Gender'].value_counts()

# Frequency table for KP281 Gender
frequency_tables['Gender'] = pd.DataFrame({
    'Gender': kp281_gender_counts.index,
    'Count': kp281_gender_counts.values,
    'Percentage': (kp281_gender_counts.values / kp281_gender_counts.sum()) * 100
})

# Product - Age (20s and 30s)
age_20s_30s_count = data[(data['Age'] >= 20) & (data['Age'] < 40)].shape[0]
frequency_tables['Age'] = pd.DataFrame({
    'Age Range': ['20-39', 'Other'],
    'Count': [age_20s_30s_count, total_customers - age_20s_30s_count],
    'Percentage': [(age_20s_30s_count / total_customers) * 100, ((total_customers - age_20s_30s_count) / total_customers) * 100]
})

# Product - Income
low_income_count = data[data['Income'] <= 60000].shape[0]
high_income_count = data[data['Income'] > 100000].shape[0]
frequency_tables['Income'] = pd.DataFrame({
    'Income Level': ['Low Income (<= 60K)', 'High Income (> 100K)', 'Other'],
    'Count': [low_income_count, high_income_count, total_customers - low_income_count - high_income_count],
    'Percentage': [
        (low_income_count / total_customers) * 100,
        (high_income_count / total_customers) * 100,
        ((total_customers - low_income_count - high_income_count) / total_customers) * 100
    ]
})

# Product - Fitness
fitness5_count = data[data['Fitness'] == 5].shape[0]
fitness5_kp781_count = data[(data['Fitness'] == 5) & (data['Product'] == 'KP781')].shape[0]
frequency_tables['Fitness'] = pd.DataFrame({
    'Fitness Level': ['Fitness 5', 'Other'],
    'Count': [fitness5_count, total_customers - fitness5_count],
    'Percentage': [
        (fitness5_count / total_customers) * 100,
        ((total_customers - fitness5_count) / total_customers) * 100
    ]
})

# Product - Marital Status
marital_status_counts = data['MaritalStatus'].value_counts()
frequency_tables['MaritalStatus'] = pd.DataFrame({
    'Marital Status': marital_status_counts.index,
    'Count': marital_status_counts.values,
    'Percentage': (marital_status_counts.values / total_customers) * 100
})

# Display all frequency tables
for key, table in frequency_tables.items():
    print(f"Frequency Table for {key}:\n")
    print(table)
    print("\n")
```
